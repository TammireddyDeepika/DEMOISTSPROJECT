```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
#step1:Simple neural network (denoiser)
class SimpleDenoiser(nn.Module):
    def __init__(self):
        super().__init__()
        self.model=nn.Sequential(
            nn.Linear(784,512),
            nn.ReLU(),
            nn.Linear(512,784),
        )
    def forward(self,x):
        return self.model(x)

#step2:Initialize model
model=SimpleDenoiser()


#step3:Create pure noise image
noise=torch.randn(1,784)#28x28 image flattened

#step4:Reverse diffusion(denoising)
steps=10
image=noise.clone()

for step in range(steps):
     image=model(image)#remove noise gradually
#step5:Reshape and display
final_image=image.view(28,28).detach()

plt.imshow(final_image,cmap='gray')
plt.title('Image Generated from Noise(Diffusion Demo)')
plt.axis("off")
plt.show()

```
