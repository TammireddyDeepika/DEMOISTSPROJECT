```python
# Import necessary modules
from transformers import pipeline, AutoTokenizer


def load_model():
    """
    Loads and returns the text generation pipeline
    using a pre-trained GPT-2 model.
    """
    tokenizer = AutoTokenizer.from_pretrained("gpt2")
    generator = pipeline(
        "text-generation",
        model="gpt2",
        tokenizer=tokenizer
    )
    return generator, tokenizer


def generate_ai_text(generator, tokenizer, prompt):
    """
    Generates text based on the user's prompt.
    Handles empty input safely.
    """
    if not prompt.strip():
        return "Error: Please enter a valid text prompt."

    response = generator(
        prompt,
        max_new_tokens=85,
        truncation=True,
        pad_token_id=tokenizer.eos_token_id
    )

    return response[0]["generated_text"]


def main():
    """
    Main execution function
    """
    print("=== Generative AI Text Generator ===")

    # Read user input
    user_prompt = input("Enter your text prompt: ")

    # Load AI model
    generator, tokenizer = load_model()

    # Generate AI output
    output = generate_ai_text(generator, tokenizer, user_prompt)

    # Display result
    print("\nGenerated Output:\n")
    print(output)


# Program entry point
if __name__ == "__main__":
    main()
```
